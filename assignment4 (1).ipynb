```python
#5 Drop all rows with N/A,NA,na values in Num_Bedrooms

#df['NUM_BEDROOMS'].dropna() 
list1=['n/a','NA','na',np.NaN]
df[df.NUM_BEDROOMS.isin(list1) == False]
```

```python
#6 Replace the NaN and String entries in Num_Bath with previous entry


for i in range(0,len(df.NUM_BEDROOMS)):
    if (pd.isna(df.NUM_BEDROOMS[i])):
        df.NUM_BEDROOMS[i]=df.NUM_BEDROOMS[i-1]
    elif  df.NUM_BEDROOMS[i]=='na':
        df.NUM_BEDROOMS[i]=df.NUM_BEDROOMS[i-1]
    else:
       pass
        

print(df)
```

```python
#7. Replace the empty values in “SQ_FT” with the mean of the all entries.  


# throwing error when using .mean() directly

#pd.to_numeric with the argument errors='coerce', which turns values to NaN if it can't convert it to numeric. 
A=df.PID.sum()
B=df.PRICE.sum()
C=pd.to_numeric(df['SQ_FT'],errors='coerce').sum()
D=pd.to_numeric(df['NUM_BATH'],errors='coerce').sum()
E=pd.to_numeric(df['NUM_BEDROOMS'],errors='coerce').sum()

mean=(A+B+C+D+E)/df.size

for i in range(0,len(df.SQ_FT)):
    if (pd.isna(df.SQ_FT[i])) or (df.SQ_FT[i]=='--'):
        df.SQ_FT[i]=mean

        

print(df)
```

```python
#13 Display the state having highest average price per square foot of housing area. Hint: Average of (TotalPrice/Sq_Ft)


avg=df.PRICE / pd.to_numeric(df.SQ_FT,errors='coerce')

df['price per sqr ft']=avg

temp=df.groupby('STATE')['price per sqr ft'].mean()
print(temp,"\n\n")


print("state having highest average price per square foot of housing area is : " , temp.idxmax())
```
